# Processing


```
import pandas as pd 

file_handler = open("train.csv", "r") 
file_handler2 = open("test.csv", "r") 

train = pd.read_csv(file_handler, sep = ",") 
test = pd.read_csv(file_handler2, sep = ",") 

file_handler.close() 
file_handler2.close() 
```

```
import tensorflow.keras as keras


y_train = train.iloc[:,0]

x_train = keras.utils.normalize(train.iloc[:,1:], 
                               axis=1)
```

```
train = x_train.join(y_train)
```

```
from keras.models import Model

model = Sequential()

model.add(Dense(784, activation='relu'))
model.add(Dense(1024, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile('adam', 
              'mean_squared_error',
              metrics=['accuracy'])

```

```
from sklearn.preprocessing import LabelEncoder
import pandas as pd

encoder =  LabelEncoder()



y1 = encoder.fit_transform(train['label'])

Y = pd.get_dummies(y1)

train.drop(columns = "label", 
          inplace=True) 


train = Y.join(train)



train.head()
```
